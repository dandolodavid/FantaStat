```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from selenium import webdriver
import time
from tqdm import tqdm
```

```python
url = 'https://www.fantacalcio.it/quotazioni-fantacalcio'
#html = requests.get(url).text
```

```python
chromepath = "selenium/webdriver/chrome/chromedriver.exe"
browser = webdriver.Chrome(chromepath)
browser.get(url)
time.sleep(10)
browser.find_element_by_xpath('//*[@id="qc-cmp2-ui"]/div[3]/div/button[3]').click()
time.sleep(10)
html = browser.page_source
soup = BeautifulSoup(html, 'lxml')
```

```python
goalkeeper_url = []
for player in tqdm(soup.findAll('table',{'class':'table tabledt responsiveTb simple stripe row-border order-column dttable sfull tbq dataTable DTFC_Cloned'})[1].findAll('a')):
    goalkeeper_url.append(player.get('href'))
browser.close()
```

```python
#go to D
browser = webdriver.Chrome(chromepath)
browser.get(url)
time.sleep(10)
browser.find_element_by_xpath('//*[@id="qc-cmp2-ui"]/div[3]/div/button[3]').click()
time.sleep(10)
browser.find_element_by_xpath('//*[@id="qttabs"]/li[2]/a').click()
time.sleep(10)
soup = BeautifulSoup( browser.find_element_by_xpath('//*[@id="tbdifensori_wrapper"]/div[2]/div[3]/div[2]/div/table').get_attribute('innerHTML'), 'lxml')
def_url = []
for player in tqdm(soup.findAll('a')):
    def_url.append(player.get('href'))
browser.close()
```

```python
#go to C
browser = webdriver.Chrome(chromepath)
browser.get(url)
time.sleep(10)
browser.find_element_by_xpath('//*[@id="qc-cmp2-ui"]/div[3]/div/button[3]').click()
time.sleep(10)
browser.find_element_by_xpath('//*[@id="qttabs"]/li[3]/a').click()
time.sleep(10)
soup = BeautifulSoup( browser.find_element_by_xpath('//*[@id="tbcentrocampisti_wrapper"]/div[2]/div[3]/div[2]/div/table').get_attribute('innerHTML'), 'lxml')
cen_url = []
for player in tqdm(soup.findAll('a')):
    cen_url.append(player.get('href'))
browser.close()
```

```python
#go to A
browser = webdriver.Chrome(chromepath)
browser.get(url)
time.sleep(10)
browser.find_element_by_xpath('//*[@id="qc-cmp2-ui"]/div[3]/div/button[3]').click()
time.sleep(10)
browser.find_element_by_xpath('//*[@id="qttabs"]/li[4]/a').click()
time.sleep(10)
soup = BeautifulSoup( browser.find_element_by_xpath('//*[@id="tbattaccanti_wrapper"]/div[2]/div[3]/div[2]/div/table').get_attribute('innerHTML'), 'lxml')
att_url = []
for player in tqdm(soup.findAll('a')):
    att_url.append(player.get('href'))
browser.close()
```

```python
full_url = goalkeeper_url + def_url + cen_url + att_url
```
